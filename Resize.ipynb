### Resize
This notebook resizes all the pictures for the specified set to the dimensions specified by the size variable. It's a simple resize, the original aspect ratio of the image is not taken into account

Folders 'data/resized/train', 'data/resized/validation', 'data/resized/test' need to be created prior to the execution of this notebook

```python
import os
import pandas as pd

from skimage import io
from skimage.transform import resize
from skimage.util import img_as_ubyte

from joblib import Parallel, delayed
```

```python
size = (300, 300)

threads = 6
subset = 'validation'

folder = "."+ os.sep + "data" + os.sep
file =  subset + os.sep + subset + ".csv"

df = pd.read_csv(folder + file,sep=",")
df.head()
```

```python
def resize_pic(loadpath, savepath, fname):
    image = io.imread(loadpath + fname)
    image_resized = resize(image, size, anti_aliasing=True)
    image_resized = img_as_ubyte(image_resized)
    io.imsave(fname = savepath + fname, arr = image_resized)
    return image_resized
```

```python
loadpath = folder + subset + os.sep
savepath = folder + 'resized' + os.sep + subset + os.sep
```

```python
Parallel(n_jobs=threads)(delayed(resize_pic)(loadpath, savepath, str(i) + '.jpg') for i in df['mediaid']);
```
